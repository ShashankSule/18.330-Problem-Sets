# A continuously differentiable function with compact support

```julia
using LinearAlgebra
using PyPlot
```

```julia
function f(x,ϵ)
    if abs(x) > ϵ 
        return 0 
    else
        return 1/ϵ
    end
end

function simpson_rule(f, a, b, N)
    Δ = (b - a)/N
    h = Δ/2
    
    I = h*f(a)/3 + h*f(b)/3
    for n=1:N-1
        I += h*( 2*f(a + 2*n*h) + 4*f(a+ (2*n-1)*h))/3
    end
    I += 4*h*f(a+ (2*N-1)*h)/3
    
    return I
end

kernel(x,y,ϵ) = (1/ϵ)*exp(-((x-y)/ϵ)^2)
```

```julia
ϵ = 1/16
y = LinRange(-π, π, 1000)
functionvals = zeros(1000)
for i in 1:1000
    tempkernel(x) = f(x,ϵ)*kernel(x,y[i],ϵ)
    functionvals[i] = simpson_rule(tempkernel, -π, π, 1000)
end
```

```julia
plot(y,functionvals)
```

```julia
function digswith2(n)
    sum = 0.0
    for i in 1:n 
        digs = digits(i)
        for j in 1:length(digs)
            if digs[j] == 2
                sum = sum + 1.0
            end
        end
    end
    return Int64(trunc(sum))
end
```

```julia
nums = [i for i in 1:100]
plot(nums, digswith2.(nums), "o-")
plot(nums, nums, "x-")
```
